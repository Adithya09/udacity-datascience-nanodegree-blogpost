## Question 1: Remote work and job satisfaction
#### What are the main indicators of job satisfaction and can we use these relevant indicators to accurately predict job satisfaction for a given employee in the technology field?

```python
q1_columns = ['HomeRemote', 'JobSatisfaction', 'Professional', 'Salary', 'Country', 'CompanyType', 'CompanySize']
q1_df = df[q1_columns]
```

```python
q1_df.head()
```

```python
# Check for missing values and data types in the selected columns
print(q1_df.info())
print(q1_df.isnull().sum())
```

### Filtering data to only include entries with salary data

```python
q1_df_with_salary = q1_df.dropna(subset=['Salary'])
```

```python
print(q1_df_with_salary.isnull().sum())
```

```python
# Since less than 5% of the subset is null, let us remove these rows
```

```python
q1_df_clean = q1_df_with_salary.dropna()
```

```python
print(q1_df_clean.isnull().sum())
```

```python
# Checking for outliers in salary
plt.figure(figsize=(10, 6))
sns.boxplot(x=q1_df['Salary'])
plt.title('Boxplot of Salary')
plt.xlabel('Salary')
plt.show()
```

```python
q1_df_clean.loc[:, 'JobSatisfactionCategory'] = pd.cut(
    q1_df_clean['JobSatisfaction'],
    bins=[0, 3, 7, 10],
    labels=['0-3', '4-7', '8-10'],
    right=False
)
```

```python
# Remote Work Status vs. Job Satisfaction
plt.figure(figsize=(12, 6))
sns.countplot(x='HomeRemote', hue='JobSatisfactionCategory', data=q1_df_clean)
plt.title('Job Satisfaction Categories by Remote Work Status')
plt.xlabel('Remote Work Status')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Job Satisfaction Score')
plt.tight_layout()
plt.show()
```
